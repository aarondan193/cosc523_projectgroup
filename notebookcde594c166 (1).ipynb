```python
#train_data = train_data.dropna()
#test_data = test_data.dropna()
train_data['Age'] = train_data['Age'].fillna(train_data['Age'].median())
test_data['Age'] = test_data['Age'].fillna(test_data['Age'].median())

test_data['Fare'] = test_data['Fare'].fillna(test_data['Fare'].median())
```

```python
#categorical variables
train_data['cabin_adv'] = train_data.Cabin.apply(lambda x: str(x)[0])
test_data['cabin_adv'] = test_data.Cabin.apply(lambda x: str(x)[0])
```

```python
train_data.head()
```

```python
train_data['ticket_letters'] = train_data.Ticket.apply(lambda x: ''.join(x.split(' ')[:-1]).replace('.','').replace('/','').lower() if len(x.split(' ')[:-1]) >0 else 0)
test_data['ticket_letters'] = test_data.Ticket.apply(lambda x: ''.join(x.split(' ')[:-1]).replace('.','').replace('/','').lower() if len(x.split(' ')[:-1]) >0 else 0)
```

```python
from sklearn.ensemble import RandomForestClassifier

y = train_data["Survived"]
train_data = train_data.drop(['Survived'], axis=1)

#features = ["Pclass", "Sex", "SibSp", "Parch"]
#X = pd.get_dummies(train_data[features])
#X_test = pd.get_dummies(test_data[features])

train_data.drop(['Embarked','cabin_adv','Name','Ticket','Cabin',], axis=1, inplace=True)
test_data.drop(['Embarked','cabin_adv','Name','Ticket','Cabin',], axis=1, inplace=True)

features = ['Sex','ticket_letters']
X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])

train_data = pd.merge(train_data, X, left_index=True, right_index=True)
test_data = pd.merge(test_data, X, left_index=True, right_index=True)

train_data.drop(['Sex','ticket_letters'], axis=1, inplace=True)
test_data.drop(['Sex','ticket_letters'], axis=1, inplace=True)

model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(train_data, y)
predictions = model.predict(test_data)

output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```
